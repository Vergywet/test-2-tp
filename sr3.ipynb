```python
sns.set_style("darkgrid")
#sns.set_context("talk")
sns.displot(data['OBJECTID'])
plt.title('Distribution graph for TOBJECT_ID')
plt.show() 
```

```python
output = data.head(12) 

plt.figure(figsize=(12, 6))
sns.barplot(x='NAME_OF_SE', y='EST_HHC',  data=output)
plt.title('The estimated household per stree name')
plt.xlabel('The street names')
plt.ylabel('Estimated household count per street')
plt.xticks(rotation=45)
plt.show()
```

```python
hold = data.nlargest(8,'OBJECTID')

name = hold['NAME_OF_SE']
output = hold['OBJECTID']

plt.figure(figsize = (10,8))
plt.pie(output,labels = name,autopct= '%1.1f%%',startangle=140)
plt.title('Pie chart for Name OF SE and Objectid')
plt.show()
```

```python
data =  np.random.normal(loc=20.0,scale=5.0,size = 365)
plt.hist(data,bins=20,color = "Purple",alpha= 0.7)
plt.title('Histograph of  EST_HHC vs OBJECTID')
plt.xlabel('EST_HHC')
plt.ylabel('OBJECTID')
plt.show()
```

```python
est_hhc = [0,1,2,3,4,5,6]

sns.kdeplot(est_hhc,shade=True)
plt.title('A kde graph for Clean sheets')
plt.xlabel('EST_HHC')
plt.show()
```
